```python
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands  = mp.solutions.hands
hands = mp_hands.Hands(model_complexity=0, min_detection_confidence=0.5, min_tracking_confidence=0.5)
```

```python
cap = cv2.VideoCapture(0)
samples = []
while cap.isOpened():
    success, image = cap.read()
    if not success:
        print("Ignoring empty camera frame.")
        continue

    # To improve performance, optionally mark the image as not writeable to
    # pass by reference.
    image.flags.writeable = False
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(image)
    # Draw the hand annotations on the image.
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            features = []
            for id, lm in enumerate(hand_landmarks.landmark):
                print(f"{id} : {lm}")
                mp_drawing.draw_landmarks(
                    image,
                    hand_landmarks,
                    mp_hands.HAND_CONNECTIONS,
                    mp_drawing_styles.get_default_hand_landmarks_style(),
                    mp_drawing_styles.get_default_hand_connections_style())
                features.append([int(lm.x),int(lm.y)])
            samples.append(features)
        print(np.array(samples).shape)
    # Flip the image horizontally for a selfie-view display.
    cv2.imshow('MediaPipe Hands', cv2.flip(image, 1))
    if cv2.waitKey(5) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
class Extractor:
    """ Class for extracting features of the input frame """
    def __init__(self, extractor = 'hand'):
        self.extractor = extractor
        self.mp_drawing = mp.solutions.drawing_utils
        self.mp_drawing_styles = mp.solutions.drawing_styles
        self.mp_hands  = mp.solutions.hands
        self.hands = mp_hands.Hands(model_complexity=0, min_detection_confidence=0.5, min_tracking_confidence=0.5)
    
    def extract(self, inputData, features_size = None):
        inputData.flags.writeable = False
        inputData = cv2.cvtColor(inputData, cv2.COLOR_BGR2RGB)
        results = self.hands.process(inputData)
        # Draw the hand annotations on the image.
        inputData.flags.writeable = True
        inputData = cv2.cvtColor(inputData, cv2.COLOR_RGB2BGR)
        if results.multi_hand_landmarks:
            features = np.array([])
            for hand_landmarks in results.multi_hand_landmarks:
                for id, lm in enumerate(hand_landmarks.landmark):
                    if lm.x and lm.y:
                        if features.size == 0:
                            features = np.array([int(lm.x*inputData.shape[1]), int(lm.y*inputData.shape[0])])
                        else:
                            features = np.hstack((features, np.array([int(lm.x*inputData.shape[1]), int(lm.y*inputData.shape[0])])))
            if not features_size:
                features_size = features.shape[0]
            features = np.reshape(features, (1, features_size))
            #print(features.shape)
            return features
```

```python
class Snap:
    def __init__(self, inputType = 'image', 
                 activator = mouseActivator('left'), 
                 outputIndicator = None, 
                 featureExtractor = None, 
                 labels = 'input', 
                 showData = False):
        self.inputType = inputType
        self.activator = activator
        self.featureExtractor = Extractor(featureExtractor)
        self.dataset = np.array([])
        if labels == 'input':
            self.labeler = Labeler('classID')
        if outputIndicator:
            self.outputIndicator = Indicator(outputIndicator)
            
    def addData(self, inputStream):
        self.inputStream = inputStream
        while self.inputStream.isOpened():
            success, image = self.inputStream.read()
            image = cv2.flip(image, 1)
            self.imgShape = (image.shape[0], image.shape[1])
            if not success:
                print("Ignoring empty camera frame.")
                continue
                
            if self.activator.isActive():
                sequence = self.addSequence()
                if sequence.any():
                    print(sequence.shape)
                label = self.labeler.getCurrentLabel()
                label_arr = label * np.ones((sequence.shape[0],1))
                sequence = np.hstack((sequence, label_arr))
                
                print(label_arr.shape)
                if self.dataset.size > 0:
                    self.dataset = np.vstack((self.dataset, sequence))
                else:
                    self.dataset = sequence
                #print(f'sequence = {sequence}, label = {label}')
            cv2.imshow('MediaPipe Hands',image)
            if cv2.waitKey(5) & 0xFF == 27:
                break
        inputStream.release()
        cv2.destroyAllWindows()
    
    
    def addSequence(self):
        sequence = np.array([])
        while self.activator.isActive() and self.inputStream.isOpened():
            success, image = self.inputStream.read()
            image = cv2.flip(image, 1)
            if not success:
                print("Ignoring empty camera frame.")
                continue
            
            sample = self.featureExtractor.extract(image)
            
            try:
                if sample.size > 0 and sequence.size > 0:
                    sequence = np.vstack((sequence,sample))
                else:
                    sequence = sample
                cv2.imshow('MediaPipe Hands',image)
                
            except:
                print("no sample")
            
            if cv2.waitKey(5) & 0xFF == 27 or not self.activator.isActive():
                    break
        if sequence.size > 0:
            return sequence
            
        
    def saveDataset(self, path = 'report', datasetName = 'out', featureList = None):
        if featureList:
            if len(featureList) == self.dataset.shape[0] - 1:
                self.featureList = featureList
        else:
            self.featureList = [i for i in range(self.dataset.shape[1] - 1)]
            
        columnsList = self.featureList + ['Label']
        
        df = pd.DataFrame(np.array(self.dataset), columns = columnsList)
        
        if not os.path.exists(path):
            os.makedirs(path, exist_ok=True)
            df.to_csv(path+"/"+datasetName+".csv")
        else:
            if(os.path.exists(path+"/"+datasetName+".csv") and os.path.isfile(path+"/"+datasetName+".csv")):
                os.remove(path+"/"+datasetName+".csv")
                print("Deleted")
            df.to_csv(path+"/"+datasetName+".csv")
    
    def printDataset(self):
        print(np.array(self.dataset).shape)
        columnsList = [i for i in range(42)] + ['Label']
        df = pd.DataFrame(np.array(self.dataset), columns = columnsList)
        print(df)
        
    def plotData(self):
        label_list = np.unique(self.dataset[1:,-1])
        #for row in self.dataset:
            
        print(label_list)
        fig, ax = plt.subplots()
        data = self.dataset[:,:-1]
        
        plt.xlim(0, 600)
        plt.ylim(0, 600)
        print(data.shape)
        colorIdx = 0
        for row in data[:,24:26]:
            plt.plot([point for idx, point in enumerate(row) if idx%2 == 0], 
                     [point for idx, point in enumerate(row) if idx%2 != 0],'o' ,color = str(1/(2**(colorIdx/len(data)))))
            colorIdx+=1
        #plt.invert_xaxis()
        plt.gca().invert_yaxis()
        plt.show()
```

```python
import datetime

async def display_date():
    loop = asyncio.get_running_loop()
    end_time = loop.time() + 5.0
    while True:
        print(datetime.datetime.now())
        if (loop.time() + 1.0) >= end_time:
            break
        await asyncio.sleep(1)
        
async def Hello():
    loop = asyncio.get_running_loop()
    end_time = loop.time() + 5.0
    while True:
        print('Hello')
        if (loop.time() + 1.0) >= end_time:
            break
        await asyncio.sleep(3)
        

L = await asyncio.gather(
        display_date(),
        Hello()
    )
print(L)
```
