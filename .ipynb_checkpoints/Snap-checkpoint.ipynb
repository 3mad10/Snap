```python
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands  = mp.solutions.hands
hands = mp_hands.Hands(model_complexity=0, min_detection_confidence=0.5, min_tracking_confidence=0.5)
```

```python
cap = cv2.VideoCapture(0)
samples = []
while cap.isOpened():
    success, image = cap.read()
    if not success:
        print("Ignoring empty camera frame.")
        continue

    # To improve performance, optionally mark the image as not writeable to
    # pass by reference.
    image.flags.writeable = False
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(image)
    # Draw the hand annotations on the image.
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            features = []
            for id, lm in enumerate(hand_landmarks.landmark):
                print(f"{id} : {lm}")
                mp_drawing.draw_landmarks(
                    image,
                    hand_landmarks,
                    mp_hands.HAND_CONNECTIONS,
                    mp_drawing_styles.get_default_hand_landmarks_style(),
                    mp_drawing_styles.get_default_hand_connections_style())
                features.append([int(lm.x),int(lm.y)])
            samples.append(features)
        print(np.array(samples).shape)
    # Flip the image horizontally for a selfie-view display.
    cv2.imshow('MediaPipe Hands', cv2.flip(image, 1))
    if cv2.waitKey(5) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
class Extractor:
    """ Class for extracting features of the input frame """
    def __init__(self, extractor = 'hand', features_size = 0, feature_dimensions = 1):
        self.extractor = extractor
        self.mp_drawing = mp.solutions.drawing_utils
        self.mp_drawing_styles = mp.solutions.drawing_styles
        self.mp_hands  = mp.solutions.hands
        self.hands = mp_hands.Hands(model_complexity=0, min_detection_confidence=0.5, min_tracking_confidence=0.5)
        self.features_size = features_size
        self.feature_dimensions = feature_dimensions
    def extract(self, inputData):
        inputData.flags.writeable = False
        inputData = cv2.cvtColor(inputData, cv2.COLOR_BGR2RGB)
        results = self.hands.process(inputData)
        # Draw the hand annotations on the image.
        inputData.flags.writeable = True
        inputData = cv2.cvtColor(inputData, cv2.COLOR_RGB2BGR)
        if results.multi_hand_landmarks:
            features = np.array([])
            for hand_landmarks in results.multi_hand_landmarks:
                for id, lm in enumerate(hand_landmarks.landmark):
                    if lm.x and lm.y:
                        if features.size == 0:
                            features = np.array([int(lm.x*inputData.shape[1]), int(lm.y*inputData.shape[0])])
                        else:
                            features = np.hstack((features, np.array([int(lm.x*inputData.shape[1]), int(lm.y*inputData.shape[0])])))
            if not self.features_size:
                self.features_size = features.shape[0]
                
            try:
                features = np.reshape(features, (self.feature_dimensions, self.features_size))
            except ValueError:
                return None

            return features
    
    def getFeatureSize(self):
        return self.features_size
    
    def setFeatureSize(self, feature_size = 0):
        if feature_size<=0:
            self.features_size = feature_size
        else:
            print("Feature size can not be less than or equal to zero")
    
    
```

```python
class Snap:
    def __init__(self, inputType = 'image', 
                 activator = mouseActivator('left'), 
                 outputIndicator = None, 
                 featureExtractor = None, 
                 labels = 'input', 
                 showData = False):
        self.inputType = inputType
        self.activator = activator
        self.sequenceID = 0
        self.latestActiveState = False
        self.featureExtractor = Extractor(featureExtractor)
        self.dataset = pd.DataFrame()
        if labels == 'input':
            self.labeler = Labeler(["a","b","c","d"])
        if outputIndicator:
            self.outputIndicator = Indicator(outputIndicator)
            
    def addData(self, inputStream):
        self.inputStream = inputStream
        
        while self.inputStream.isOpened():
            success, image = self.inputStream.read()
            image = cv2.flip(image, 1)
            self.imgShape = (image.shape[0], image.shape[1])
            if self.activator.isActive() and not self.latestActiveState:
                self.sequenceID = self.sequenceID + 1
            if not success:
                print("Ignoring empty camera frame.")
                continue
            if self.activator.isActive():
                self.addSample(image)
            self.latestActiveState = self.activator.isActive()
            cv2.imshow('MediaPipe Hands',image)
            if cv2.waitKey(5) & 0xFF == 27:
                break
        inputStream.release()
        cv2.destroyAllWindows()
    
    def addSample(self, sample):
        sampleFeature = self.featureExtractor.extract(sample)
        sample = np.append(sampleFeature, [self.labeler.getCurrentLabel(), self.sequenceID])
        try:
            if sampleFeature.any():
                self.dataset = self.dataset.append(pd.Series(sample), ignore_index = True)
        except:
            print("Features could not be extracted from the input sample")
            
    def saveDataset(self, path = 'report', datasetName = 'out', featureList = None):
        if featureList:
            if len(featureList) == self.dataset.shape[0] - 1:
                self.featureList = featureList
        else:
            self.featureList = [i for i in range(self.dataset.shape[1] - 2)]
            
        columnsList = self.featureList + ['Label','Sequence No']
        
        df = pd.DataFrame(np.array(self.dataset), columns = columnsList)
        
        if not os.path.exists(path):
            os.makedirs(path, exist_ok=True)
            df.to_csv(path+"/"+datasetName+".csv")
        else:
            if(os.path.exists(path+"/"+datasetName+".csv") and os.path.isfile(path+"/"+datasetName+".csv")):
                os.remove(path+"/"+datasetName+".csv")
                print("Deleted")
            df.to_csv(path+"/"+datasetName+".csv")
    
    def printDataset(self):
        print(self.dataset)
        #columnsList = [i for i in range(42)] + ['Label']
        #df = pd.DataFrame(np.array(self.dataset), columns = columnsList)
        
    def plotData(self, sliceOfInterest = [24,25]):
        label_list = np.unique(self.dataset.iloc[1:,-2])
        #for row in self.dataset:
            
        #print(label_list)
        fig, ax = plt.subplots()
        #print(self.dataset.iloc[:,:-2])
        data = self.dataset.iloc[:,:-2]
        
        dataOfInterest = data.iloc[:,sliceOfInterest]
        plt.xlim(0, 600)
        plt.ylim(0, 600)
        #print(data.shape)
        colorIdx = 0
        #print(data.iloc[:,24:26])
        for index, row in dataOfInterest.iterrows():
            #print(row)
            #for idx, point in enumerate(row):
            #    print(f"index : {idx}, point : {point}")
            #print([int(point) for idx, point in enumerate(row) if idx%2 == 0])
            #print([int(point) for idx, point in enumerate(row) if idx%2 != 0])
            plt.plot([int(point) for idx, point in enumerate(row) if idx%2 == 0], 
                     [int(point) for idx, point in enumerate(row) if idx%2 != 0],'o' ,color = str(1/(2**(colorIdx/len(data)))))
            colorIdx+=1
        plt.gca().invert_yaxis()
        plt.show()
```

```python
import datetime

async def display_date():
    loop = asyncio.get_running_loop()
    end_time = loop.time() + 5.0
    while True:
        print(datetime.datetime.now())
        if (loop.time() + 1.0) >= end_time:
            break
        await asyncio.sleep(1)
        
async def Hello():
    loop = asyncio.get_running_loop()
    end_time = loop.time() + 5.0
    while True:
        print('Hello')
        if (loop.time() + 1.0) >= end_time:
            break
        await asyncio.sleep(3)
        

L = await asyncio.gather(
        display_date(),
        Hello()
    )
print(L)
```
